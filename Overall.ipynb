```python
# Extract total visitors for the 'Total' data series
total_visitors = df[df['Data Series'] == 'Total'].iloc[0, 1:]

# Exclude columns with '2024'
columns_to_exclude = [col for col in total_visitors.index if '2024' in col]
filtered_total_visitors = total_visitors.drop(columns_to_exclude)

years_months = [(col.split()[0], col.split()[1]) for col in filtered_total_visitors.index if col.split()[0].isdigit()]
filtered_total_visitors.index = pd.MultiIndex.from_tuples(years_months, names=["Year", "Month"])

quarters = {
    "Jan": "Q4", "Feb": "Q4", "Mar": "Q1",
    "Apr": "Q1", "May": "Q1", "Jun": "Q2",
    "Jul": "Q2", "Aug": "Q2", "Sep": "Q3",
    "Oct": "Q3", "Nov": "Q3", "Dec": "Q4"
}

filtered_total_visitors = filtered_total_visitors.reset_index()
filtered_total_visitors['Quarter'] = filtered_total_visitors['Month'].map(quarters)
filtered_total_visitors = filtered_total_visitors.groupby(['Year', 'Quarter']).sum()
quarterly_total_visitors = filtered_total_visitors.groupby('Quarter').sum(numeric_only=True)
quarterly_total_visitors = quarterly_total_visitors.astype(float)

max_value = quarterly_total_visitors.max()[0]
min_value = quarterly_total_visitors.min()[0]
max_quarter = quarterly_total_visitors.idxmax()[0]
min_quarter = quarterly_total_visitors.idxmin()[0]

colors = ['#8d89c0', '#8d89c0', '#8d89c0', '#8d89c0']

# Plotting the bar chart
plt.figure(figsize=(10, 5))
bars = plt.bar(quarterly_total_visitors.index, quarterly_total_visitors.values.flatten(), color=colors)

plt.xlabel('Quarter', fontsize=14)
plt.ylabel('Total Number of Visitors', fontsize=14)
plt.title('Figure 4  Total Number of Visitors by Quarter from 2008 to 2023', fontsize=15, pad=30)

# Annotate the bars with values and percentages

total_visitors_sum = quarterly_total_visitors.sum()[0]
for bar in bars:
    height = bar.get_height()
    percentage = (height / total_visitors_sum) * 100
    plt.annotate(f'{height:.0f}\n{percentage:.1f}%', 
                 xy=(bar.get_x() + bar.get_width() / 2, height/2), 
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=12, color='black')

# Highlight max and min bars with text
for bar in bars:
    if bar.get_height() == max_value:
        plt.annotate('Max', 
                     xy=(bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                     xytext=(0, 3),  # 3 points vertical offset
                     textcoords="offset points",
                     ha='center', va='bottom', fontsize=12, color='red')
    elif bar.get_height() == min_value:
        plt.annotate('Min', 
                     xy=(bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                     xytext=(0, 3),  # 3 points vertical offset
                     textcoords="offset points",
                     ha='center', va='bottom', fontsize=12, color='green')

plt.tight_layout()
plt.show()
```

```python
# Figure8: average stay of length and quarter
import pandas as pd
import matplotlib.pyplot as plt

# Filter the data for "Average Length Of Stay (Day)"
average_stay = df1[df1['Data Series'] == 'Average Length Of Stay (Day)'].iloc[0, 1:]

# Exclude columns with '2020-2024' in them
columns_to_exclude = [col for col in average_stay.index if any(year in col for year in ['2020', '2021', '2022', '2023', '2024'])]
filtered_average_stay = average_stay.drop(columns_to_exclude)

# Convert to numeric, handling errors by coercing them to NaN
filtered_average_stay = filtered_average_stay.apply(pd.to_numeric, errors='coerce')

# Extract years and months
years = [int(col.split()[0]) for col in filtered_average_stay.index]
months = [col.split()[1] for col in filtered_average_stay.index]
filtered_average_stay.index = pd.MultiIndex.from_arrays([years, months], names=["Year", "Month"])

# Sort the MultiIndex
filtered_average_stay = filtered_average_stay.sort_index()

# Filter data for 2008 to 2019
filtered_average_stay = filtered_average_stay.loc[2008:2019]

# Define quarters
month_to_quarter = {'Jan': 'Q1', 'Feb': 'Q1', 'Mar': 'Q1',
                    'Apr': 'Q2', 'May': 'Q2', 'Jun': 'Q2',
                    'Jul': 'Q3', 'Aug': 'Q3', 'Sep': 'Q3',
                    'Oct': 'Q4', 'Nov': 'Q4', 'Dec': 'Q4'}

# Add quarter information
quarters = [month_to_quarter[month] for month in filtered_average_stay.index.get_level_values('Month')]
filtered_average_stay.index = pd.MultiIndex.from_arrays([filtered_average_stay.index.get_level_values('Year'), quarters], names=["Year", "Quarter"])

# Group by quarter and calculate the mean for each quarter across years
quarterly_average_stay = filtered_average_stay.groupby('Quarter').mean()
quarterly_average_stay = quarterly_average_stay.astype(float)

# Ensure the quarters are ordered from Q1 to Q4
quarter_order = ['Q1', 'Q2', 'Q3', 'Q4']
quarterly_average_stay = quarterly_average_stay.reindex(quarter_order)

# Determine max and min values for annotations
max_value = quarterly_average_stay.max()
min_value = quarterly_average_stay.min()
max_quarter = quarterly_average_stay.idxmax()
min_quarter = quarterly_average_stay.idxmin()

# Plot as a bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(quarterly_average_stay.index, quarterly_average_stay.values, color='#b9b9da')
plt.xlabel('Quarter', fontsize=14)
plt.ylabel('Average Stay Length (Days)', fontsize=14)
plt.title('T2-Figure4  Average Stay Length by Quarter (2008 to 2019)', fontsize=16, pad=20)
plt.xticks(rotation=0)
plt.ticklabel_format(style='plain', axis='y')

# Annotate max and min values
plt.text(max_quarter, max_value, f'Max: {max_value:.2f}', ha='center', va='bottom', fontsize=10, color='red')
plt.text(min_quarter, min_value, f'Min: {min_value:.2f}', ha='center', va='bottom', fontsize=10, color='green')

plt.tight_layout()
plt.show()
```
